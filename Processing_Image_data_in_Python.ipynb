## **Image Data Processing**

**Converting Image into Numpy Array**

```python
#Store image as mpimg file to convert into array
img = mpimg.imread('/content/bts.jpg')
type(img)
img.shape
```

```python
#display image 
img_plot = plt.imshow(img)
plt.show()
```

**Resizing Image Using Pillow**

```python
img = Image.open('/content/bts.jpg')
img_resized = img.resize((200,200))
#save file to the current directory to find it's shape
img_resized.save('bts_reshaped.jpg')
```

```python
#display image in numpy array
img_resized = mpimg.imread('/content/bts_reshaped.jpg')
img_resized.shape
img_resized_plot = plt.imshow(img_resized)
plt.show()
```

**Converting RGB to GreyScale Image**

```python
img = cv2.imread('/content/bts.jpg')
```

```python
grey_image = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
```

```python
cv2_imshow(grey_image)
```

```python
cv2.imwrite('bts_grayscale.jpg', grey_image)
```
